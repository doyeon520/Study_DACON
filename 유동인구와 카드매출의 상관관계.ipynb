```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc("font",family="Malgun Gothic")
```

# 2019년 주간 유동인구의합과 주간 카드매출의 합의 단순회귀분석

```python
df_2019 = b.drop('weeks',axis=1)
df_2019
```

```python
X = df_2019["유동인구"]
y = df_2019["USE_AMT"]
plt.plot(X, y, 'o')
plt.show()
```

```python
line_fitter = LinearRegression()
line_fitter.fit(X.values.reshape(-1,1), y)
```

```python
#기울기
line_fitter.coef_
```

```python
#절편
line_fitter.intercept_
```

```python
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.title('2019')
plt.xlabel('유동인구')
plt.ylabel('카드매출')
plt.show()
```
