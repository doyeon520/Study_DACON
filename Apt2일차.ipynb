```python
#park['dong']에대해 park_count 딕셔너리로 만들기
dic=park.set_index('dong').to_dict()
dic['park_count']
```

### 시계열 데이터는 관측 값들이 시간에 따라 변하는 자료를 말한다. (ex. 주가, 주가 수익률, 변동성 등)

시계열 자료는 인접된 값들이 서로 완전히 독립적 이라고는 할 수 없고 상호 종속적 관계에 있는데, 현재는 과거에 종속된다. 완전히 독립적 관계의 시계열 자료도 존재한다. (ex : 백색 잡음, 주가의 수익률)

시계열 자료는 일반적인 통계 이론이나 통계적 분석 방법으로는 다루기 어렵고, 시계열 분석 (Time Series Analysis) 방법을 사용한다. 주식의 일별 시세, 원.달러 시세, 변동성 시세 등과 같은 대부분의 금융 관련 자료는 시계열 자료의 특성을 가진다.

금융 시계열 자료는 월별 출생자수, 태양의 월별 흑점수 등과 같은 통상의 시계열 자료 보다 복잡한 구조를 갖는다.

시계열은 안정(stationary) 시계열과 불안정(non-stationary) 시계열 두 가지로 구분되는데, 안정 시계열은 데이터의 어떤 구간에서도 평균값과 분산(데이터의 흩어짐 정도)이 일정한 경우를 말합니다. 이를 정상 시계열이라고도 합니다.
시계열 분석을 위해서 안정 시계열로 변환해야 하는데 RNN, LSTM 과 같은 인공지능 모델은 비선형적 모델이라고 할 수 있기때문에 안정시계열로 변환할 필요가 없습니다.
